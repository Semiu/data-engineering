```python
# the library

import psycopg2
```

```python
# Connect to the database

try:
    conn = psycopg2.connect("dbname=udacity")
except psycopg2.Error as e:
    print("Error connecting to the DB")
    print(e)

try:
    cur = conn.cursor()
except psycopg2.Error as e:
    print("Error getting cursor")
    print(e)
conn.set_session(autocommit = True)
```

```python
# Create all the tables

try:
    cur.execute("CREATE TABLE IF NOT EXISTS album_library (album_id int, album_name varchar, artist_id int, year int);")
except psycopg2.Error as e:
    print("Error creating table")
    print(e)

try:
    cur.execute("CREATE TABLE IF NOT EXISTS artist_library (artist_id int, artist_name varchar);")
except psycopg2.Error as e:
    print("Error creating table")
    print(e)
    
try:
    cur.execute("CREATE TABLE IF NOT EXISTS song_library (song_id int, album_id int, song_name varchar);")
except psycopg2.Error as e:
    print("Error creating table")
    print(e)

try:
    cur.execute("CREATE TABLE IF NOT EXISTS song_length (song_id int, song_length int);")
except psycopg2.Error as e:
    print("Error creating table")
    print(e)

# Insert into tables
try:
    cur.execute("INSERT INTO song_length (song_id, song_length) \
    VALUES (%s, %s)", \
               (1, 163))
except psycopg2.Error as e:
    print("Error inserting")
    print(e)
    
try:
    cur.execute("INSERT INTO song_length (song_id, song_length) \
    VALUES (%s, %s)", \
               (2, 137))
except psycopg2.Error as e:
    print("Error inserting")
    print(e)

try:
    cur.execute("INSERT INTO song_length (song_id, song_length) \
    VALUES (%s, %s)", \
               (3, 145))
except psycopg2.Error as e:
    print("Error inserting")
    print(e)
    
    
try:
    cur.execute("INSERT INTO song_length (song_id, song_length) \
    VALUES (%s, %s)", \
               (4, 240))
except psycopg2.Error as e:
    print("Error inserting")
    print(e)
    
# So many code left unwritten - they are just INSERT statements for all the tables that made the 3NF



# Remember to drop the tables used for experiementing and close the cursor and the connection
```
