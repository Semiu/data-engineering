```python
# Function to convert the CSV file to a dataframe. It takes the path to the data file as parameter and returns a dataframe of the data

def csv_to_df(path):
  datadf = pd.read_csv(path, low_memory=False)
  return datadf
```

```python
# This function takes a text description as a parameter and returns three lists which categorise the content of the text description into device
# patient, and any other one

def extract_keyword_corpus(text_desr):

  device_problem = []
  device_problem_2 = []
  patient_problem = []
  patient_problem_2 = []
  other_problem = []
  
  global final_device_problem_1, final_device_problem_2, final_device_problem, final_patient_problem_1, final_patient_problem_2, final_patient_problem, final_other_problem_keyphrases
  
  text_list = str(text_desr).split('.')
  
  for text in text_list:
    if bool(re.search("device", text, re.IGNORECASE)):
      device_problem.append(text)

    ##
    elif bool(re.search("reported", text, re.IGNORECASE)):
      device_problem_2.append(text)

    if bool(re.search("customer", text, re.IGNORECASE)):
      patient_problem_2.append(text)
    ##

    elif bool(re.search("patient", text, re.IGNORECASE)):
      patient_problem.append(text)
      
    else:
      other_problem.append(text)
  
  if len(device_problem) > 0:
    device_problem_keyphrases_1 = [d_problem.split('device')[1] for d_problem in device_problem]
    device_problem_keyphrases_2 = [d_problem.split('device')[0] for d_problem in device_problem]

    device_problem_keyphrases_3 = [device_problem_keyphrases_1.split(',') for device_problem_keyphrases_1 in device_problem_keyphrases_1]
    device_problem_keyphrases_4 = [device_problem_keyphrases_2.split(',') for device_problem_keyphrases_2 in device_problem_keyphrases_2]

    final_device_problem_keyphrases_1 = [item.strip() for sublist in device_problem_keyphrases_3 for item in sublist]
    final_device_problem_keyphrases_2 = [item.strip() for sublist in device_problem_keyphrases_4 for item in sublist]

    final_device_problem_1 = [*final_device_problem_keyphrases_1, *final_device_problem_keyphrases_2]
  
  elif len(device_problem) == 0:
    final_device_problem_1 = device_problem
  
  if len(device_problem_2) > 0:
    device_problem_keyphrases_5 = [d_problem.split('reported')[1] for d_problem in device_problem_2]
    device_problem_keyphrases_6 = [d_problem.split('reported')[0] for d_problem in device_problem_2]

    device_problem_keyphrases_7 = [device_problem_keyphrases_5.split(',') for device_problem_keyphrases_5 in device_problem_keyphrases_5]
    device_problem_keyphrases_8 = [device_problem_keyphrases_6.split(',') for device_problem_keyphrases_6 in device_problem_keyphrases_6]
    
    final_device_problem_keyphrases_3 = [item.strip() for sublist in device_problem_keyphrases_7 for item in sublist]
    final_device_problem_keyphrases_4 = [item.strip() for sublist in device_problem_keyphrases_8 for item in sublist]

    final_device_problem_2 = [*final_device_problem_keyphrases_3, *final_device_problem_keyphrases_4]

  elif len(device_problem_2) == 0:
    final_device_problem_2 = device_problem_2
  
  if len(patient_problem) > 0:
    patient_problem_keyphrases_1 = [p_problem.split('patient')[1] for p_problem in patient_problem]
    patient_problem_keyphrases_2 = [p_problem.split('patient')[0] for p_problem in patient_problem]

    patient_problem_keyphrases_3 = [patient_problem_keyphrases_1.split(',') for patient_problem_keyphrases_1 in patient_problem_keyphrases_1]
    patient_problem_keyphrases_4 = [patient_problem_keyphrases_2.split(',') for patient_problem_keyphrases_2 in patient_problem_keyphrases_2]

    final_patient_problem_keyphrases_1 = [item.strip() for sublist in patient_problem_keyphrases_3 for item in sublist if item != ""]
    final_patient_problem_keyphrases_2 = [item.strip() for sublist in patient_problem_keyphrases_4 for item in sublist if item != ""]

    final_patient_problem_1 = [*final_patient_problem_keyphrases_1, *final_patient_problem_keyphrases_2]
  
  elif len(patient_problem) == 0:
    final_patient_problem_1  = patient_problem

  if len(patient_problem_2) > 0:
    patient_problem_keyphrases_5 = [p_problem.split('customer')[1] for p_problem in patient_problem_2]
    patient_problem_keyphrases_6 = [p_problem.split('customer')[0] for p_problem in patient_problem_2]

    patient_problem_keyphrases_7 = [patient_problem_keyphrases_5.split(',') for patient_problem_keyphrases_5 in patient_problem_keyphrases_5]
    patient_problem_keyphrases_8 = [patient_problem_keyphrases_6.split(',') for patient_problem_keyphrases_6 in patient_problem_keyphrases_6]

    final_patient_problem_keyphrases_3 = [item.strip() for sublist in patient_problem_keyphrases_7 for item in sublist if item != ""]
    final_patient_problem_keyphrases_4 = [item.strip() for sublist in patient_problem_keyphrases_8 for item in sublist if item != ""]

    final_patient_problem_2 = [*final_patient_problem_keyphrases_3, *final_patient_problem_keyphrases_4]

  elif len(patient_problem_2) > 0:
    final_patient_problem_2 = patient_problem_2

  if len(other_problem) > 0:
    other_problem_keyphrases = [o_problem.split(',') for o_problem in other_problem]
    final_other_problem_keyphrases = [item.strip() for sublist in other_problem_keyphrases for item in sublist if item != ""]
  
  elif len(other_problem) == 0:
    final_other_problem_keyphrases = other_problem
  
  final_device_problem = [*final_device_problem_1, *final_device_problem_2]
  final_patient_problem = [*final_patient_problem_1, *final_patient_problem_2]

  return final_device_problem, final_patient_problem, final_other_problem_keyphrases
```

```python
# Customized stop words. This would be expanded to include words to be removed form the final output

stopwords = ('there', 'was', 'the', 'customer', 'therefore')
```

```python
# A tokenizer function, but modified to exclude Lemmatizer and NLTK tokenizer. Instead, split the sentence by white space to get each word of a 
# sentence as token. It also does include the digits

def modified_tokenizer(sentence):
  
  eliminate_words = [',',"'",'.','(',')','{','}',':','e.g',';', '@', '{}', '()', '(^a-z)(0-9$)'] # This can also be increased
  sentence = sentence.lower() # downcase
  tokens = [t for t in sentence.split(" ")]
  tokens = [t for t in tokens if len(t) > 2 or len(t)==1]   # remove short words, they're probably not useful
  tokens = [t for t in tokens if t not in eliminate_words] # Remove those weird characters in the eliminate_words list
  tokens = [t for t in tokens if t not in stopwords] # remove stopwords
  digits = [t for t in tokens if any(c.isdigit() for c in t)] # get the digits seperately (probably these are the scores)
  
  return " ".join(tokens), digits
```

```python
# This function takes the text description as a parameter, passes it to the extract_keyword_corpus function.
# The returned values from the extract_keyword_corpus function are further processed to get a dictionary data structure for the required features
# This function returns a dictionary where the key-values are the six columns to be populated in the dataframe

def extract_keywords_dict(text_desr):
  
  keyword_corpus = extract_keyword_corpus(text_desr) #The extract_keyword_corpus function is called here
  
  keyword_corpus_device = keyword_corpus[0]
  keyword_corpus_patient = keyword_corpus[1]
  keyword_corpus_other = keyword_corpus[2]
  
  keywords_dict = {}
  
  device_keywords = []
  device_score = []
  patient_keywords = []
  patient_score = []
  other_keywords = []
  other_score = []
  
  if len(keyword_corpus_device) > 0:
    for sent_d in keyword_corpus_device:
      sentence_tokens_d = modified_tokenizer(sent_d) # The tokenizer function is called here
      if len(sentence_tokens_d[0]) >= 1:
        device_keywords.append(sentence_tokens_d[0])
      if len(sentence_tokens_d[1]) >= 1:
        device_score.append(sentence_tokens_d[1])
  
  if len(keyword_corpus_patient) > 0:
    for sent_p in keyword_corpus_patient:
      sentence_tokens_p = modified_tokenizer(sent_p) # The tokenizer function is called here
      if len(sentence_tokens_p[0]) >= 1:
        patient_keywords.append(sentence_tokens_p[0])
      if len(sentence_tokens_p[1]) >= 1:
        patient_score.append(sentence_tokens_p[1])
 
  if len(keyword_corpus_other) > 0:
    for sent_o in keyword_corpus_other:
      sentence_tokens_o = modified_tokenizer(sent_o) # The tokenizer function is called here
      if len(sentence_tokens_o[0]) >= 1:
        other_keywords.append(sentence_tokens_o[0])
      if len(sentence_tokens_o[1]) >= 1:
        other_score.append(sentence_tokens_o[1])

  if any(isinstance(x,list) for x in device_keywords):
    device_keywords = [item.strip() for sublist in device_keywords for item in sublist if item != ""]

  if any(isinstance(x,list) for x in device_score):
    device_score = [item.strip() for sublist in device_score for item in sublist if item != ""]

  if any(isinstance(x,list) for x in patient_keywords):
    patient_keywords = [item.strip() for sublist in patient_keywords for item in sublist if item != ""]

  if any(isinstance(x,list) for x in patient_score):
    patient_score = [item.strip() for sublist in patient_score for item in sublist if item != ""]

  if any(isinstance(x,list) for x in other_keywords):
    other_keywords = [item.strip() for sublist in other_keywords for item in sublist if item != ""]

  if any(isinstance(x,list) for x in other_score ):
    other_score = [item.strip() for sublist in other_score for item in sublist if item != ""]

  keywords_dict["keywords (Device Problems)"] = ",".join(device_keywords)
  keywords_dict["keywords (Device Problems)_score"] = ",".join(device_score)
  keywords_dict["keywords (Patient Problems)"] = ",".join(patient_keywords)
  keywords_dict["keywords (Patient Problems)_score"] = ",".join(patient_score)
  keywords_dict["keywords others (interesting ones)"] = ",".join(other_keywords)
  keywords_dict["keywords others (interesting ones)_score"] = ",".join(other_score)

  return keywords_dict
```

```python
# Improved process data frame function

def process_data(datadf):
  
  for (idx, row) in datadf.iterrows():
    text_descr = row['TEXT DESCRIPTION']
    keywords_to_use = extract_keywords_dict(text_descr)
    datadf.loc[idx,['keywords (Device Problems)','keywords (Device Problems)_score', 'keywords (Patient Problems)','keywords (Patient Problems)_score', 'keywords others (interesting ones)','keywords others (interesting ones)_score']] = [keywords_to_use.get("keywords (Device Problems)", "Not available"), keywords_to_use.get("keywords (Device Problems)_score", "Not available"),
                                keywords_to_use.get("keywords (Patient Problems)", "Not available"), keywords_to_use.get("keywords (Patient Problems)_score", "Not available"),
                                keywords_to_use.get("keywords others (interesting ones)", "Not available"), keywords_to_use.get("keywords others (interesting ones)_score", "Not available")]
  return datadf
```
